# Cleaning Data

**lets clean the data buy:**

> Removing Unnecessary columns 

> Removing Null values

> droping duplicates

```python
#Removing unnecessary columns
data.drop(['url','menu_item','dish_liked','phone','reviews_list'], axis=1, inplace=True)
data.head(10)
```

```python
#Removing null values
data.dropna(how = 'any', inplace=True)
data.shape
```

```python
#Removing duplicate values
data = data[data.duplicated() == False]
data.shape
```

```python
#Assigning proper data type to the columns
data.info()
```

```python
data['approx_cost(for two people)'] = data['approx_cost(for two people)'].str.replace(',','')
data['approx_cost_for_2'] = data['approx_cost(for two people)'].astype(int)
```

```python
data['approx_cost_for_2'].value_counts()
```

## Question 1

**Which type of restaurants are providing table booking facility in the bengaluru city?**

```python
sns.countplot(x=data['book_table'],hue= data['listed_in(type)'])
fig=plt.gcf()
plt.title('Type of restaurants providing table booking facility')
```

```python
book_no = data[data['book_table'] == 'No']
book_no =book_no['listed_in(type)'].value_counts().sort_index()
book_no
```

```python
book_yes = data[data['book_table'] == 'Yes']
book_yes = book_yes['listed_in(type)'].value_counts().sort_index()
book_yes
```

```python
diff = book_no - book_yes
diff
```

### Answer

> Not all the restaurants provide facilities of booking table cafes, delivery, Desserts and Dine-Out these restaurants prefer not to have table booking

> But the restaurants types like Buffet, Drinks & nightlife and Pubs & bars prefer to have a table book before you visit these restaurants.

## Question 2

**Which area in the city have more number of restaurants located?**

```python
fig=plt.figure(figsize=(40,200))
locationCount=data['location'].value_counts().sort_values(ascending=True)

locationCount.plot(kind='barh',fontsize=50)
plt.ylabel('Location Names',fontsize = 70)
plt.title('Restaurent Count graoh Vs Location',fontsize= 100)

for i in range(len(locationCount)):
    plt.text(i+locationCount[i],i,locationCount[i],fontsize=50)
```

```python
locationCount.nlargest(10)
```

```python
locationCount.nsmallest(10)
```

## Question 3

**What are the cheapest and the costly prices of the restaurants in different parts ofthe city?**

```python
temp = data.groupby('location')['approx_cost_for_2'].agg([('Cmin','min'), ('Cmax','max')])
ax = temp.plot(figsize=(25,8), grid=True)
ax.set_xticks(range(len(temp)))
ax.set_xticklabels(temp.index, rotation=90)
ax.set_xlabel('Location', fontsize=20)
ax.set_ylabel('Cost', fontsize=20)
ax.set_title('Title:- Cost range', fontsize=20)
plt.show()
```

```python
temp['Cmin'].nsmallest(5)
```

```python
temp['Cmax'].nlargest(5)
```
